# アレイを使ったデータ処理

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 軸の目盛りになるアレイを用意
points = np.arange(-10,10,0.1)
```

```python
# グリッドを作る
dx,dy=np.meshgrid(points,points)
```

```python
dx
```

```python
dy
```

```python
plt.imshow(dx)
```

```python
plt.imshow(dy)
```

```python
z = (np.sin(dx) + np.sin(dy))
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title("Plot for sin(x)+ sin(y)")
```

```python
A = np.array([1,2,3,4])
```

```python
B = np.array([1000,2000,3000,4000])
```

```python
condition = np.array([True, True, False, False])
```

```python
answer = [(a if cond else b) for a,b,cond in zip(A,B,condition)]
answer
```

```python
answer2 = np.where(condition, A, B)
answer2
```
